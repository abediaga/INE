```python
df_B = df_bizkaia.copy()
df_A = df_araba.copy()
df_G = df_gipuzkoa.copy()

df_B = df_B.set_index(["Año","Mes"])
df_A = df_A.set_index(["Año","Mes"])
df_G = df_G.set_index(["Año","Mes"])

d = {}
d['Bizkaia'] = df_B
d['Gipuzkoa'] = df_G
d['Araba'] = df_A

df = pd.DataFrame()
df = pd.concat(d, axis=1)
```

```python
display(df[["Bizkaia","Gipuzkoa"]])
df.loc[:, (("Bizkaia","Gipuzkoa","Araba"), ("Total Constituidas"))]
display(df.loc[2019:, (("Bizkaia","Gipuzkoa","Araba"), ("Total Constituidas"))])
display(df.loc[2019:, (("Bizkaia","Gipuzkoa","Araba"), ("Total Disueltas"))])
df.loc[((2018,2019,2020), (1)), (("Bizkaia","Gipuzkoa","Araba"), ("Total Constituidas"))]
display(df.loc[(df.index.get_level_values(0).to_list(), (1)), (("Bizkaia","Gipuzkoa","Araba"), ("Total Constituidas"))])
display(df.loc[(df.index.get_level_values(0).to_list(), (2)), (list(set(list(zip(*df.columns))[0])), ("Total Constituidas","Total Disueltas"))])

anos = df.index.get_level_values(0).to_list()
meses = range(1,13)
provincias = list(set(list(zip(*df.columns))[0]))
columnas = ["Total Constituidas","Total Disueltas"]
display(df.loc[(anos, meses), (provincias, columnas)])

data = {'Año':  df.index.get_level_values(0).to_list(),
       'Mes':  df.index.get_level_values(1).to_list(),
       'Bizkaia Total Constituidas':  df["Bizkaia"]["Total Constituidas"].to_list(),
       'Gipuzkoa Total Constituidas': df["Gipuzkoa"]["Total Constituidas"].to_list(),
       'Araba Total Constituidas': df["Araba"]["Total Constituidas"].to_list()}
dff = pd.DataFrame (data, columns = ["Año","Mes","Bizkaia Total Constituidas","Gipuzkoa Total Constituidas","Araba Total Constituidas"])
display(dff.set_index(["Año","Mes"]))
```
